### 1. Task Manager Desktop GUI Version

```python
import ipywidgets as widgets
from IPython.display import display, clear_output

task_list = []
output = widgets.Output()

text_input = widgets.Text(placeholder='Enter a task', description='Task:')
add_button = widgets.Button(description='Add Task', button_style='success')
clear_button = widgets.Button(description='Clear Completed', button_style='warning')

def render_tasks():
    with output:
        clear_output()
        if not task_list:
            print("No tasks yet!")
        for i, task in enumerate(task_list):
            print(f"{i+1}. {task}")

def on_add_clicked(b):
    if text_input.value:
        task_list.append(text_input.value)
        text_input.value = ''
        render_tasks()

def on_clear_clicked(b):
    global task_list
    task_list = [t for t in task_list if '(Done)' not in t]
    render_tasks()

add_button.on_click(on_add_clicked)
clear_button.on_click(on_clear_clicked)

display(widgets.HBox([text_input, add_button, clear_button]), output)
render_tasks()
```

### 2. Task Manager with Advanced Features

```python
import ipywidgets as widgets
from IPython.display import display, clear_output
import json
import os

DB_FILE = 'tasks.json'

def load_tasks():
    if os.path.exists(DB_FILE):
        with open(DB_FILE, 'r') as f:
            return json.load(f)
    return []

def save_tasks(tasks):
    with open(DB_FILE, 'w') as f:
        json.dump(tasks, f)

tasks = load_tasks()
output = widgets.Output()

task_entry = widgets.Text(placeholder='What needs to be done?', description='Task:')
priority = widgets.Dropdown(options=['Low', 'Medium', 'High'], value='Medium', description='Priority:')
add_btn = widgets.Button(description='Add Task', button_style='success', icon='plus')
clear_btn = widgets.Button(description='Clear All', button_style='danger', icon='trash')

def refresh_view():
    with output:
        clear_output()
        if not tasks:
            print("Your task list is empty.")
        for i, t in enumerate(tasks):
            print(f"{i+1}. [{t['priority']}] {t['name']}")

def handle_add(b):
    if task_entry.value:
        tasks.append({'name': task_entry.value, 'priority': priority.value})
        save_tasks(tasks)
        task_entry.value = ''
        refresh_view()

def handle_clear(b):
    tasks.clear()
    save_tasks(tasks)
    refresh_view()

add_btn.on_click(handle_add)
clear_btn.on_click(handle_clear)

ui = widgets.VBox([
    widgets.HBox([task_entry, priority]),
    widgets.HBox([add_btn, clear_btn]),
    output
])

display(ui)
refresh_view()
```

```python
from fpdf import FPDF
import ipywidgets as widgets
from IPython.display import display, FileLink
import os

def generate_pdf(text, filename="generated_document.pdf"):
    pdf = FPDF()
    pdf.add_page()
    pdf.set_font("Arial", size=12)
    # Multi-cell allows for line breaks
    pdf.multi_cell(0, 10, txt=text)
    pdf.output(filename)
    return filename

# UI Elements
text_area = widgets.Textarea(
    placeholder='Type your PDF content here...',
    description='Content:',
    layout={'height': '200px', 'width': '90%'}
)
filename_input = widgets.Text(value='my_document.pdf', description='Filename:')
gen_button = widgets.Button(description='Generate PDF', button_style='info')
output_link = widgets.Output()

def on_button_clicked(b):
    with output_link:
        output_link.clear_output()
        fname = filename_input.value if filename_input.value.endswith('.pdf') else filename_input.value + '.pdf'
        generate_pdf(text_area.value, fname)
        print(f"Success! PDF saved as {fname}")
        display(FileLink(fname))

gen_button.on_click(on_button_clicked)

display(text_area, filename_input, gen_button, output_link)
```

### 4. Email News App (News API + Email)

```python
import requests
import smtplib
from email.mime.text import MIMEText

def get_news(api_key, query='technology'):
    url = f'https://newsapi.org/v2/everything?q={query}&apiKey={api_key}'
    response = requests.get(url)
    articles = response.json().get('articles', [])[:5]
    content = 'Latest News:\n\n'
    for art in articles:
        content += f"- {art['title']}\n{art['url']}\n\n"
    return content

def send_email(user, password, recipient, body):
    msg = MIMEText(body)
    msg['Subject'] = 'Daily News Update'
    msg['From'] = user
    msg['To'] = recipient
    with smtplib.SMTP_SSL('smtp.gmail.com', 465) as server:
        server.login(user, password)
        server.send_message(msg)
```

### 7. Create a Weather API

```python
import requests

def get_current_weather(city, api_key):
    url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        main = data['main']
        weather = data['weather'][0]
        return {
            "temp": main['temp'],
            "humidity": main['humidity'],
            "description": weather['description']
        }
    return None
```

### 10. Webcam Motion Alert App

```python
from IPython.display import display, Javascript
from google.colab.output import eval_js
from base64 import b64decode

def take_photo(filename='photo.jpg', quality=0.8):
  js = Javascript('''
    async function takePhoto(quality) {
      const div = document.createElement('div');
      const capture = document.createElement('button');
      capture.textContent = 'Capture';
      div.appendChild(capture);
      const video = document.createElement('video');
      video.style.display = 'block';
      const stream = await navigator.mediaDevices.getUserMedia({video: true});
      document.body.appendChild(div);
      div.appendChild(video);
      video.srcObject = stream;
      await video.play();
      google.colab.output.setIframeHeight(document.documentElement.scrollHeight, true);
      await new Promise((resolve) => capture.onclick = resolve);
      const canvas = document.createElement('canvas');
      canvas.width = video.videoWidth;
      canvas.height = video.videoHeight;
      canvas.getContext('2d').drawImage(video, 0, 0);
      stream.getVideoTracks()[0].stop();
      div.remove();
      return canvas.toDataURL('image/jpeg', quality);
    }
    ''')
  display(js)
  data = eval_js('takePhoto({})'.format(quality))
  binary = b64decode(data.split(',')[1])
  with open(filename, 'wb') as f:
    f.write(binary)
  return filename
```

### 11. Web Scrape Music Events

```python
import requests
from bs4 import BeautifulSoup

def scrape_events(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    events = []
    for item in soup.select('.event-item'):
        title = item.select_one('.title').text
        date = item.select_one('.date').text
        events.append({'title': title, 'date': date})
    return events
```

### 12. Hotel Booking System with OOP

```python
class Room:
    def __init__(self, room_no, price):
        self.room_no = room_no
        self.price = price
        self.is_booked = False

class Hotel:
    def __init__(self, name):
        self.name = name
        self.rooms = []
    def add_room(self, room):
        self.rooms.append(room)
    def book_room(self, room_no):
        for room in self.rooms:
            if room.room_no == room_no and not room.is_booked:
                room.is_booked = True
                return f"Room {room_no} booked successfully!"
        return "Room not available."
```

### 13. Mario Game in OOP Style

```python
class Entity:
    def __init__(self, x, y):
        self.x = x
        self.y = y

class Mario(Entity):
    def __init__(self, x, y):
        super().__init__(x, y)
        self.velocity_y = 0
        self.is_jumping = False
    def jump(self):
        if not self.is_jumping:
            self.velocity_y = -10
            self.is_jumping = True
    def update(self):
        self.y += self.velocity_y
        if self.y < 0:
            self.velocity_y += 1
```

### 20. Movie Recommendation System

```python
import pandas as pd

def get_recommendations(movie_title, df):
    if movie_title not in df['title'].values:
        return "Movie not found."
    genre = df[df['title'] == movie_title]['genre'].iloc[0]
    recommendations = df[df['genre'] == genre]['title'].tolist()
    return [m for m in recommendations if m != movie_title]
```
